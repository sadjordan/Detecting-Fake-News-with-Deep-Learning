# Dataset Cleaning

## Tasks:
- Merge datasets with train, test, validation splits into 1 dataset
- Verify label is consistent across datasets (0 for false; 1 for true)
- Standardise data by removing unneeded columns (id, URLs, etc.)
- Dataset merging
    - Split off train set from processed datasets and merge them to form a big train set
    - Keep test sets seperate
    - If needed, split validation set from train set

```python
import pandas as pd
import os
```

```python
base_path = "../../Datasets"
```

## Merging datasets which are split into various subsets

```python
#Dataset 1: fake-news-detection-dataset-English

ds1_path = os.path.join(base_path, "fake-news-detection-dataset-English") 


df1_train = pd.read_csv(os.path.join(ds1_path, "train.csv"))
df1_test = pd.read_csv(os.path.join(ds1_path, "test.csv"))
df1_val = pd.read_csv(os.path.join(ds1_path, "validation.csv"))

# Merge
df_ds1_merged = pd.concat([df1_train, df1_test, df1_val], ignore_index=True)
print(f"Successfully merged. Final shape: {df_ds1_merged.shape}")

output_path = os.path.join(ds1_path, "merged.csv")
df_ds1_merged.to_csv(output_path, index=False)
```

```python
#Dataset 2: aadyasingh55/fake-news-classification

ds2_path = os.path.join(base_path, "aadyasingh55/fake-news-classification/versions/1")


df2_train = pd.read_csv(os.path.join(ds2_path, "train (2).csv"), sep=';')
df2_test = pd.read_csv(os.path.join(ds2_path, "test (1).csv"), sep=';')
df2_eval = pd.read_csv(os.path.join(ds2_path, "evaluation.csv"), sep=';')


df_ds2_merged = pd.concat([df2_train, df2_test, df2_eval], ignore_index=True)
print(f"Successfully merged. Final shape: {df_ds2_merged.shape}")

output_path = os.path.join(ds2_path, "merged.csv")
df_ds2_merged.to_csv(output_path, index=False)
```
